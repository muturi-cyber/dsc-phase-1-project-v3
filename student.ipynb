# Introduction.

Our company is expanding into new industries and Specifically,our interest is in purchasing and operating airplanes for both commercial and private enterprises.Our objectives are:one, to identify  potential risks of  the aircraft business .two, to determine which aircrafts are the lowest risk for the company to start with.

```python
# setting up for data analysis and visualization.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# importing a csv file into a Panda DataFrame.

df = pd.read_csv('Aviation_Data.csv', low_memory=False)

# to preview column types and also check the data type of individual column.

df.info()
```

```python
# Displaying the first few rows to sample data,inspect the data structure and column names.

df.head()
```

```python
# Displaying the models in our data and their counts

df['Model'].value_counts()
```

```python
# Displaying the make and their counts.

df['Make'].value_counts()
```

```python
# cleaning The Data.

# Removing the missing values in the column 'Make'

df.dropna(subset=['Make'])
```

```python
# removing missing values in the column 'Total.fatal.Injuries

df.dropna(subset=['Total.Fatal.Injuries'],inplace=True)
df['Total.Fatal.Injuries']
```

```python
# grouping by make and sum the Total fatal injuries.

df.groupby('Make')['Total.Fatal.Injuries'].sum().sort_values(ascending=False).head(15)
```

```python
top_15_makes = df.groupby('Make')['Total.Fatal.Injuries'].sum().sort_values(ascending=False).head(15)
print(top_15_makes)
```

```python
# Plotting the bar chart.
top_15_makes.plot(kind='bar')

# Adding the title and the labels.
plt.title('top 15 makes with most fatal injuries ')
plt.ylabel('total fatal injuries')
plt.xlabel('Aircraft Makes')

# displaying the plot.
plt.show()
```

```python
# grouping by model and sum the Total fatal injuries.
df.groupby('Model')['Total.Fatal.Injuries'].sum().sort_values(ascending=False).head(15)
top_15_models = df.groupby('Model')['Total.Fatal.Injuries'].sum().sort_values(ascending=False).head(15)
print(top_15_models)
```

```python
# Plotting the bar chart.
top_15_models.plot(kind='bar')

# Adding the title and the labels.
plt.title('top 15 models with most fatal injuries ')
plt.ylabel('total fatal injuries')
plt.xlabel('Aircraft Models')

# displaying the plot.
plt.show()
```

```python
df['Aircraft.Category'].value_counts()
```

```python
df.dropna(subset=['Aircraft.Category'],inplace=True)
df['Aircraft.Category']
```

```python
df.groupby('Aircraft.Category')['Total.Fatal.Injuries'].sum().sort_values(ascending=False).head(10)
top_10_aircraft_category = df.groupby('Aircraft.Category')['Total.Fatal.Injuries'].sum().sort_values(ascending=False).head(10)
print(top_10_aircraft_category)
```

```python
top_10_aircraft_category.plot(kind='bar')
plt.title('top 10 aircraft categories with most fatal injuries ')
plt.ylabel('total fatal injuries')
plt.xlabel('Aircraft category')
plt.show()
```
